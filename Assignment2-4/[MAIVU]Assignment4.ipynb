## Problem 1

1. Generate 10,000 random numbers from *Poission distribution* having mean value of 50. Make a histogram of the values. Pay special attention to the bins parameter.
1. Calculate the mean and standard deviation values for the generated random numbers.
1. Generate 10,000 random numbers from *normal distribution* having the same mean and standard deviation as for the Poisson distribution.
1. Overlap the histograms of the generated data (Poisson and normal distribution).
1. How much do they differ? Explain why.

```python
# 1: Generate and plot a histogram of 10000 random numbers from Poisson distribution with the mean value of 50 
mean = 50
n = 10000
a_Poisson = np.random.poisson(mean, n)

bins = arange(min(a_Poisson) - 1, max(a_Poisson) + 5, 2)
plt.figure(figsize = (7, 5))
plt.hist(a_Poisson, bins, edgecolor='k')
plt.title('Histogram plot of the distribution')
plt.xlabel('Value', fontsize = 10)
plt.ylabel('Frequency', fontsize = 10)
plt.grid()
plt.show()

# 2: Calculate and print the mean and standard deviation values for the generated random numbers
print("The mean of the Poisson distribution = ", a_Poisson.mean())
print("The standard deviation of the Poission distribution = ", a_Poisson.std())
```

```python
# 3: Generate 10000 random numbers from normal distribution having the same mean and standard deviation as for the Poisson distribution
sd = a_Poisson.std()
a_Normal = np.random.normal(mean, sd, n)

# 4: Overlap the histograms of the generated data (Poisson and normal distribution)
bins = arange(min(min(a_Poisson),min(a_Normal)) - 2, max(max(a_Poisson),max(a_Normal)) + 2, 1)
plt.figure(figsize = (12, 7))
plt.hist(a_Poisson, bins, alpha = 0.5, edgecolor='k', label = 'Poisson')
plt.hist(a_Normal, bins, alpha = 0.5, edgecolor='k', label = 'Normal')
plt.title('Histogram plot of the distributions', fontsize = 20)
plt.xlabel('Value', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid()
plt.legend(loc = 'upper right', prop={'size': 15})
plt.show()
```

```python
# 5: About their differences
print('10 sample from Poisson distribution: ', a_Poisson[:5])
print('10 sample from normal distribution: ', a_Normal[:5])

a_Poisson2 = np.random.poisson(5, n)
a_Poisson3 = np.random.poisson(15, n)
a_Poisson4 = np.random.poisson(30, n)
bins = arange(min(a_Poisson2) - 2, max(max(a_Poisson),max(a_Normal)) + 2, 1)
plt.figure(figsize = (12, 7))
plt.hist(a_Poisson, bins, alpha = 0.3, edgecolor='k', label = 'Poisson')
plt.hist(a_Normal, bins, alpha = 0.3, edgecolor='k', label = 'Normal')
plt.hist(a_Poisson2, bins, alpha = 0.3, edgecolor='k', label = 'Poisson 2')
plt.hist(a_Poisson3, bins, alpha = 0.3, edgecolor='k', label = 'Poisson 3')
plt.hist(a_Poisson4, bins, alpha = 0.3, edgecolor='k', label = 'Poisson 4')
plt.title('Histogram plot of the distributions', fontsize = 20)
plt.xlabel('Value', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid()
plt.legend(loc = 'upper right', prop={'size': 15})
plt.show()
```

○ The first difference is that Poisson distribution is a discrete probability distribution; its distributions type is integer. While that of the normal distribution is float, meaning is a continuous probability distribution.<br>
○ When the mean is small, Poisson distribution is not the same as normal distribution. It skews while the normal is symmetric. When the mean is large enough, Poisson looks like normal distribution on plot.

## Problem 2

In this problem the aim is to simulate the functioning of a desktop image scanner (see Lab 4 -> Image pattern simulator). The scanner has a maximum resolution of 600 DPI (dots-per-inch) and the color-resolution of the scanner is 14 bits.

Imagine that you are scanning 2.0 cm x 2.0 cm (See the figure below) sized areas having different average brightness. The average brightness for each area is [0.001, 0.002, 0.005, 0.01, 0.02, 0.03, 0.10, 0.20, 0.50, 0.99] of the full-scale color-resolution (14-bits = 2^14).

- Write a code that simulates the scanning of the documents having different brightness.
- Calculate the mean and the standard deviation values of the brightness values for each document. 
- Plot how the signal-to-noise ratio (SNR) changes when the brightness increases (e.g. x-axis = brightess, y-axis = SNR).

```python
brightness = [0.001, 0.002, 0.005, 0.01, 0.02, 0.03, 0.10, 0.20, 0.50, 0.99]
n = int(600 * (2 / 2.54)**2)

SNR = []
for i in brightness:
    a = np.zeros(n)
    pt = int(2**14 * i)
    for j in arange(0, pt):
        temp = numpy.random.randint(n)
        a[temp] += 1 
    
    SNR.append(a.mean() / a.std())

    print("With brightness equals to", i)
    print("mean =", round(a.mean(),3))
    print("standard deviation =", round(a.std(),3))
    print("SNR =", round(a.mean() / a.std(),3), '\n')
```

```python
plt.figure(figsize = (10,5))
plot(brightness, SNR,'o:')
plt.xlabel("Brightness", fontsize = 15)
plt.ylabel("SNR", fontsize = 15)
plt.title("The signal-to-noise ratio corresponds to the brightness", fontsize = 15)
plt.grid()
plt.show()
```

## Problem 3

Find or take a photo with smooth surfaces having different lightning conditions. You could search for example [black-and-white portrait photography](https://duckduckgo.com/?q=black+and+white+portrait+photography&t=ffab&atb=v150-1&iax=images&ia=images) or [black-and-white scenery photography](https://duckduckgo.com/?q=black+and+white+scenery+photograph&t=ffab&atb=v150-1&iar=images&iax=images&ia=images).

Select parts of the images where the brightness have different intensities and calculate the mean, the standard devation and the SNR-values for those parts of the images. 

Study does the SNR vs. mean intensity value rule work also here. Make a graph (x-axis = mean intensity, y-axis = SNR) or show in a table format the calculated mean intensity values and SNR values. 

```python
import skimage.io as skio
import skimage.color as skcolor

fpath = 'http://browseideas.com/wp-content/uploads/2012/01/Stunning-Examples-of-black-and-white-Portrait-Photography-34-575x573.png'
image = skio.imread(fpath)

plt.figure(figsize=(5,5))
plt.imshow(image)
plt.show()

gray_image = skcolor.rgb2gray(image)
plt.figure(figsize=(5,5))
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
image_part = []
image_part.append(gray_image[500:600, 50:250])
image_part.append(gray_image[200:300, 250:450])
image_part.append(gray_image[350:450, 250:450])

image_mean = []
SNR = []
for i in image_part:
    print('\n\n\n')
    plt.figure(figsize=(5,5))
    plt.imshow(i, cmap='gray', interpolation='none', aspect='equal')
    plt.show()

    flat_mat = i.flatten()
    
    plt.figure()
    plt.hist(flat_mat, bins = 'auto')
    plt.title('Black and white distributions')
    plt.xlabel('Brightness (0/Black - 1/White)')
    plt.ylabel('Number of pixels')
    plt.grid()
    plt.show()

    image_mean.append(flat_mat.mean())
    SNR.append(flat_mat.mean()/flat_mat.std())
    
    print("mean =", round(flat_mat.mean(),3))
    print("standard deviation =", round(flat_mat.std(),3))
    print("SNR =", round(flat_mat.mean() / flat_mat.std(),3))
```

```python
plt.figure(figsize = (10,5))
plot(image_mean,SNR,'o:')
plt.xlabel("Mean",fontsize = 15)
plt.ylabel("SNR", fontsize = 15)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.show()
```
