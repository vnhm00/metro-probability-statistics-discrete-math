## Problem 1
Using the body girth and skeletal diameter measurements data (see the code given above) compare the distributions of the chest diameter (`che.di`) for men and women. 

- Create a graph where you have histograms of the chest diameter for both groups. Pay special attention for the bins settings.

Calculate the descriptive statistics for both groups. Compare the statistical values between the groups.
- What is the range of the values for each group?
- How much do the mean values differ between men and women?
- How much do the standard deviation values differ between men and women?
- Explain on your own words why there are similarities or differences between men and women.

```python
women_data = data[data['sex'] == 0]
men_data = data[data['sex'] == 1]
chest_min = int(min(min(women_data['che.di']), min(men_data['che.di']))) - 1
chest_max = int(max(max(women_data['che.di']), max(men_data['che.di']))) + 1
bins = arange(chest_min, chest_max, 0.7)

plt.hist(women_data['che.di'], bins, alpha = 0.5, edgecolor='k', label = 'Women', color = 'r')
plt.hist(men_data['che.di'], bins, alpha = 0.5, edgecolor='k', label = 'Men', color = 'b')
plt.legend(loc='upper right')
plt.title('Histograms of the chest diameter for women and men')
plt.xlabel('Chest diameter')
plt.ylabel('Numbers')
plt.grid()
plt.show()
```

```python
print('COMPARING CHEST DIAMETER BETWEEN WOMEN AND MEN')
print("Women data size", women_data.shape)
print("Men data size", men_data.shape, '\n')

print('The range of the values for women group: [', min(women_data['che.di']), ',' , max(women_data['che.di']), ']')
print('The range of the values for men group: [', min(men_data['che.di']), ',' , max(men_data['che.di']), ']\n')

print('The mean value for women group:', mean(women_data['che.di']))
print('The mean value for men group:', mean(men_data['che.di']))
print('The difference between women mean values and men mean values:', abs(mean(women_data['che.di']) - mean(men_data['che.di'])),'\n')

print('The sd value for women group:', std(women_data['che.di']))
print('The sd value for men group:', std(men_data['che.di']))
print('The difference between women mean values and men mean values:', abs(std(women_data['che.di']) - std(men_data['che.di'])))
```

## Problem 3
Study the elbow diameter data (elb.di). 
- Create a **density** histogram of the values both for men and women. Pay special attention to the bins settings.
- Calculate the mean and standard deviation values for both groups.
- Calculate the 95% confidence interval for the mean values. 
- Using the Â´scipy.statsÂ´ statistics normal distribution overlay the theoretical normal distribution function over the density histograms.
- How much do the density histogram and the theoretical normal distribution differ? Explain why.

```python
elb_min = int(min(min(women_data['elb.di']), min(men_data['elb.di']))) + 1
elb_max = int(max(max(women_data['elb.di']), max(men_data['elb.di']))) + 1
bins = arange(elb_min, elb_max, 0.3)

plt.hist(women_data['elb.di'], bins, alpha = 0.5, edgecolor='k', label = 'Women', color = 'r',density = True)
plt.hist(men_data['elb.di'], bins, alpha = 0.5, edgecolor='k', label = 'Men', color = 'b',density = True)

from scipy.stats import norm
women_rv = norm(loc = women_data['elb.di'].mean(), scale = women_data['elb.di'].std())
men_rv = norm(loc = men_data['elb.di'].mean(), scale = men_data['elb.di'].std())
plot(bins, women_rv.pdf(bins), alpha = 0.7, label = "Women theorical", color = "r")
plot(bins, men_rv.pdf(bins), alpha = 0.7, label = "Men theorical", color = "b")
plt.legend(loc = 'best')

plt.title('Density histograms of the elbow diameter for women and men')
plt.xlabel('Elbow diameter')
plt.ylabel('Numbers')
plt.xticks(arange(elb_min, elb_max + 1, 0.9))
plt.grid()
plt.show()
```

```python
print('COMPARING ELBOW DIAMETER BETWEEN WOMEN AND MEN')
print('The mean value for women group:', mean(women_data['elb.di']))
print('The mean value for men group:', mean(men_data['elb.di']),'\n')

print('The sd value for women group:', std(women_data['che.di']))
print('The sd value for men group:', std(men_data['che.di']),'\n')

print('95% confidence interval for the mean values of women:', mean(women_data['elb.di']) - 1.96* women_data['elb.di'].sem() ,'to', mean(women_data['elb.di']) + 1.96*women_data['elb.di'].sem())
print('95% confidence interval for the mean values of men:', mean(men_data['elb.di']) - 1.96* men_data['elb.di'].sem() ,'to', mean(men_data['elb.di']) + 1.96*men_data['elb.di'].sem())
```

The density histogram and the theoretical normal distribution are different. I believe it is because theoretical normal distribution graphs are built based on the mean and the standard deviation values.
